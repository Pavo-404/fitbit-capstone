```python
cols = ['Id','ActivityDate','TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes','Calories' ]
df = df[cols]
```

```python
df.rename(columns={'ActivityDate': 'Date'},inplace=True)
```

```python
df['TotalMimutes'] = df.VeryActiveMinutes + df.FairlyActiveMinutes + df.LightlyActiveMinutes +df.SedentaryMinutes
```

```python
df["TotlaHours"] = round(df.TotalMimutes /60).astype(int)
```

```python
plt.figure(figsize=((7,4)))
plt.hist(df.DayOfWeek,bins= 7 ,color='lightskyblue',width= 0.6)
plt.xlabel('Day of week')
plt.ylabel('Freq')
plt.title('Freq dy day ')
plt.show()
```

```python
print(df['DayOfWeek'].dtype)
print(df['DayOfWeek'].unique())

# Define a mapping from day names to numerical values
day_mapping = {'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 7}

try:
    # Map the 'DayOfWeek' column to numerical values
    df['DayOfWeek_Num'] = df['DayOfWeek'].map(day_mapping)
    
    # Plot the histogram using the numerical column
    plt.figure(figsize=(6, 5))
    plt.hist(df['DayOfWeek_Num'], bins=7, color='lightskyblue', width=0.6)
    #plt.grid(True)
    plt.show()
    
except ValueError as e:
    print("Error:", e)
```
